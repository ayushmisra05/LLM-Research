##### Load Package

```python
from langchain_openai import OpenAIEmbeddings

import os
from dotenv import load_dotenv
```

##### Set Up Environment 

```python
load_dotenv(dotenv_path = "../Key/.env")
OPENAI_API_KEY = os.environ.get("OPENAI_API_KEY")
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
```

# Create A Embedding Model

## Initialize An Engine

```python
embeddings = OpenAIEmbeddings()
```

## Build An Embedding For A Text

```python
text = "i am a king"

text_embeddings= embeddings.embed_query(text)

print(text_embeddings)
print(len(text_embeddings))
```

## Build An Embedding For A List Of Text

```python
text = ["i am a queen", 
        "i am a boy"]

text_embeddings= embeddings.embed_documents(text)

print(len(text_embeddings))
print(len(text_embeddings[0]))
```
